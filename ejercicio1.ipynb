# Transformada de Fourier de una función

```python
#Transformada de fourier
def TF(x):
    N=x.size
    n = np.arange(N)
    k = n.reshape((N,1))
    exponencial = np.exp(-2j*np.pi*k*n/N)
    return np.dot(exponencial,x)
```

```python
#Función de x a transformar

x = np.linspace(0, 2*np.pi, 500, endpoint=True)
omega= 3.
c=100.
fx=c*np.cos(2*np.pi*omega*x)
```

```python
plt.plot(x,fx)
```

```python
Transformada=TF(fx)
```

```python
plt.plot(Transformada)
```

# Código que calcule histograma de distancias para 2d y 3d de una distribución de puntos aleatorios. Para construir el histograma de distancias tienen que dividir las distancias en rangos (bins).

## En 2d:

```python
#crear la distribución random de pares

n=1000

datos=(50*np.random.random(n),50*np.random.random(n))
```

```python
plt.plot(datos[:][0],datos[:][1],'.')
```

```python
def dist(x_1,y_1,x_2,y_2):
    d=np.sqrt((x_2 - x_1)**2 + (y_2-y_1)**2)
    return d
```

```python
distaa=[]
for i in range(0,n):
    for j in range(i+1,n):
        distancia=dist(datos[0][i],datos[1][i], datos[0][j],datos[1][j])
        distaa.append(distancia)
    
```

```python
datos_n, datos_bin,m = plt.hist(distaa,bins=20)
```

## En 3d

```python
#crear la distribución random de pares

n=1000

datos_3d=(50*np.random.random(n),50*np.random.random(n),50*np.random.random(n))
```

```python
def dist_3d(x_1,y_1,z_1,x_2,y_2,z_2):
    d=np.sqrt((x_2 - x_1)**2 + (y_2-y_1)**2+(z_2-z_1)**2)
    return d
```

```python
distaa_3d=[]
for i in range(0,n):
    for j in range(i+1,n):
        distancia_3d=dist_3d(datos_3d[0][i],datos_3d[1][i], datos_3d[2][i],datos_3d[0][j],datos_3d[1][j], datos_3d[2][j])
        distaa_3d.append(distancia_3d)
```

```python
datos_n_3d, datos_bin_3d,m_3d = plt.hist(distaa_3d,bins=20)
```
